```python
# A set of physical properties of pure elements is given in the comma-separated-variable file ElementsData.csv
# Use pd.read_csv to load this data as a pandas dataframe 'ptdf'
ptdf = pd.read_csv("ElementsData.csv")
```

```python
# Check that the data has loaded correctly by displaying the first few rows of the table using ptdf.head
print(ptdf.head(5))
```

## Step 2: Calculate pairwise correlation coefficients

```python
# For a quick check for pairwise correlated features, use ptdf.corr to generate and store 
# an array of pairwise correlation coefficients. 
corr = ptdf.corr(method = 'pearson')
print(corr)
```

```python
# It can be difficult to visualize the dependencies between features using just a numerical matrix.
# Use the seaborn function sns.heatmap to create and display a heatmap of the correlation coefficients. 
coeffMap = sns.heatmap(corr)
```
